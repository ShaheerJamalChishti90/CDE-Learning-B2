### Case Study: Data Visualization (gapminder)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read data
df = pd.read_csv('./dataset/gapminder.csv')
df
```

```python
# select data
life_exp = df.life_exp
gdp_cap = df.gdp_cap
gdp_cap
```

```python

# Basic scatter plot, log scale
plt.scatter(gdp_cap, life_exp)
plt.xscale('log')
# Strings

# Add axis labels
plt.title("GDP vs Life Expectancy")
plt.xlabel("GDP-Cap Expectancy")
plt.ylabel("Life Expectancy")

# Add title

plt.show()
```

```python
pop = df.population
npp = np.array(pop)/1000000
npp
```

```python

# Basic scatter plot, log scale
plt.scatter(gdp_cap, life_exp,s=npp)
plt.xscale('log')
# Strings

# Add axis labels
plt.title("GDP vs Life Expectancy")
plt.xlabel("GDP-Cap Expectancy")
plt.ylabel("Life Expectancy")

# Add title

plt.show()

# add text in this chart
```
